```python
import pandas as pd
import csv as csv
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
#トレーニングデータ
train_df = pd.read_csv("train.csv", header=0)
train_df = train_df[["Survived","Pclass","Sex","Age","SibSp","Parch","Fare"]]
train_df_dropna =train_df.dropna()
train_df_dropna.shape
#ラベル
y_train = train_df_dropna["Survived"].values
y_train = y_train.astype(np.int64)
#特徴量
X_train_df = train_df_dropna[["Pclass","Sex","Age","SibSp","Parch","Fare"]]
X_train_df.loc[X_train_df["Sex"] == "male", "Sex"] = 0
X_train_df.loc[X_train_df["Sex"] == "female", "Sex"] = 1
X_train =X_train_df.values
X_train = X_train.astype(np.float32)
```

```python
#テストデータ
tes_df = pd.read_csv("test.csv", header=0)
passengerid_df = tes_df[["PassengerId"]]
test_df = tes_df[["Pclass","Sex","Age","SibSp","Parch","Fare"]]
print(test_df.shape)
print(passengerid_df.shape)
```

```python
#性別の符号化
test_df.loc[test_df["Sex"] == "male", "Sex"] = 0
test_df.loc[test_df["Sex"] == "female", "Sex"] = 1
```

```python
#testデータに対する欠損値補間
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='mean')
imp.fit(X_train)
X_test = imp.transform(test_df.values)
sum(np.isnan(X_test))
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
pipe_lr = make_pipeline(StandardScaler(),
                        LogisticRegression(penalty="l1",C=1,random_state=100, solver='liblinear') )
pipe_lr
```

```python
pipe_lr.fit(X_train,y_train)
y_pred =  pipe_lr.predict(X_test)
```

```python
y_pred
```

```python
passengerid = passengerid_df.values
passengerid = np.reshape(passengerid,(len(passengerid)))
result = list(zip(passengerid,y_pred))
result
```

```python
import csv
with open("predict_result_data.csv", "w") as f:
    writer = csv.writer(f, lineterminator='\n')
    writer.writerow(["PassengerId", "Survived"])
    for pid, survived in result:
        writer.writerow([pid, survived])
```
